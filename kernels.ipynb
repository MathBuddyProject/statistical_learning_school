```python
from numpy import loadtxt
import math
data = loadtxt("nn.csv", skiprows=1, delimiter=",") #Importing data file
rows = data.shape[0] #Counting the number of rows
d = len(data[0]) - 1

width = 3
matrix = [[0 for x in range(width)] for y in range(rows)]

#Storing the data from the spreadsheet in a multidimensional array
for n in range(rows):
    matrix[n] = data[n]  
    
#Checking the data for doubles and identifying them if they exist    
for n in range(rows):    
    i = n + 1
    while i < rows:
        if matrix[n][0] == matrix[i][0] and matrix[n][1] == matrix[i][1] and matrix[n][2] == matrix[i][2]:
            del matrix[n]
            rows = rows - 1
            print("A double in rows " + str(n) + " and " + str(i) + " has been found.")
        if matrix[n][0] == matrix[i][0] and matrix[n][1] == matrix[i][1] and matrix[n][2] != matrix[i][2]:
            del matrix[n]
            del matrix[i]
            rows = rows - 2
            print("A contradiction has been found where two examples are at the same point with different class in rows " + str(n) + " and " + str(i) + ".")    
        i = i + 1
        
#Taking user input for point to be classified, Kernel Rule, and smoothing factor h
print("")
x = eval(input("Enter a value for x: ")) 
y = eval(input("Enter a value for y: "))
rule = input("Enter the Kernel to be used (Gaussian, Cauchy, Triangular, Epanechnikov, Moving Window): ")
while rule != "Gaussian" and rule != "Cauchy" and rule != "Triangular" and rule != "Epanechnikov" and rule != "Moving Window":
    rule = input("Your entry was invalid. Please enter either Gaussian, Cauchy, Triangular, Epanechnikov, or Moving Window: ")
if rule == "Gaussian" or rule == "Cauchy":
    h = eval(input("Enter a value for the smoothing factor h (suggested default: 0.1): "))
elif rule == "Triangular" or rule == "Epanechnikov" or rule == "Moving Window":
    h = eval(input("Enter a value for the smoothing factor h (suggested default: 1): "))
while h <= 0:
    h = eval(input("Please enter a valid positive non-zero value for h: "))

sum = 0 #Sum of the potential function
neighbours = 0 #Number of nearest neighbours factored into the potential function

#Loop that computes the Euclidean distance of each point x from xi and plugs into kernel function, which is then plugged into the potential function        
while neighbours == 0:
    for n in range(rows):
        euclid = ((x-matrix[n][0])**2 + (y-matrix[n][1])**2)**(1/2)
        if rule == "Gaussian":
            neighbours = neighbours + 1
            kernel = math.exp(-(euclid/h)**2)
        if rule == "Cauchy":
            neighbours = neighbours + 1
            kernel = 1/(1+(euclid/h)**(d+1)) 
        if rule == "Epanechnikov":
            if euclid/h <= 1:
                neighbours = neighbours + 1
                kernel = 1-(euclid/h)**2
            else:
                kernel = 0
        if rule == "Triangular":
            if euclid/h <= 1:
                neighbours = neighbours + 1
                kernel = 1-(euclid/h)
            else:
                kernel = 0
        if rule == "Moving Window":
            if euclid <= h:
                neighbours = neighbours + 1
                kernel = 1
            else:
                kernel = 0
        sum = sum + (2*matrix[n][2]-1)*kernel   
    if neighbours == 0:
        h = eval(input("No training cases are in the domain with the chosen value of h. Please enter a new, higher value for h: "))    

#Classifying the object based on the sign of the potential function    
print("")
print("With the given h, the kernel took", neighbours, "neighbour(s) into consideration.")
if sum < 0:
    print("The object is of class 0.")
else:
    print("The object is of class 1.")
```
